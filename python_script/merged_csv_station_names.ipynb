In this notebook we will clean and process the dataset with station names in it.   
This dataset is a subset stored seperate from the master dataset in order to not skew the analysis.  
Since the number of rows in this dataset accounted for only ~1.2% of the total data, I chose to subset it and perform analysis over this dataset seperately from the master dataset.

```python
#loading dependencies

import pandas as pd
import os
import datetime as dt
import numpy as np
```

```python
# loding the dataset into a dataframe

merged_statoion_df= pd.read_csv('/Users/locco/Desktop/Google_Analytics_Course/Case_Studies/Bike_share_Data/station_merged.csv')
merged_statoion_df.drop(columns='Unnamed: 0', axis = 1, inplace = True) # dropping a duplicate index column
merged_statoion_df
```

```python
merged_statoion_df.drop(columns=['start_lat','start_lng','end_lat','end_lng', 'start_station_id', 'end_station_id'], axis = 1, inplace = True)
merged_statoion_df.info()
merged_statoion_df.isna().sum()
```

```python
# resetting index
merged_statoion_df.reset_index(drop = True, inplace= True)
```

```python
# changing column name
merged_statoion_df.rename(columns = {'member_casual':'membership_type', 'start_station_name':'origin', 'end_station_name':'destination', 'rideable_type':'bike_type'}, inplace = True)
merged_statoion_df.info()
```

```python
# checking for unique origin and destination
unique_origin = merged_statoion_df['origin'].unique()
print(len(unique_origin))
unique_destination = merged_statoion_df['destination'].unique()
print(len(unique_destination))
```
